```python
import pandas as pd
import matplotlib.pyplot as plt
# import yfinance as yf

# # Download Apple (AAPL) stock data
# data = yf.download("AAPL", start="2010-01-01", end="2025-01-01")

# # Save to CSV
# data.to_csv("AAPL.csv")

# print("✅ AAPL.csv has been downloaded successfully!")

# Load stock price data (downloaded CSV from Yahoo Finance or similar)
# Example CSV has columns: Date, Open, High, Low, Close, Volume
data = pd.read_csv("AAPL.csv")

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Plot Closing Price over time
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Close'], label="Close Price", color="blue")

# Add titles and labels
plt.title("Apple (AAPL) Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```
